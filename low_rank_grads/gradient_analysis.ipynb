```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_angles(res):
    fig,ax = plt.subplots()
    fig.set_size_inches(10, 6)
    ax.plot(res["layer_0"], label="layer_0")
    ax.plot(res["layer_1"], label="layer_1")
    ax.plot(res["layer_2"], label="layer_2")
    ax.plot(res["layer_3"], label="layer_3")
    ax.set_ylabel("Deg diff with actual gradients")
    ax.legend()
    ax2 = ax.twinx()
    ax2.plot(res["val_acc"], label="val_acc", color="gray")
    ax2.set_ylabel("Acc")
    ax2.legend()
    plt.show()
```

```python
plot_angles(pd.read_csv("cifar10_results_1.csv", index_col="epoch"))
plot_angles(pd.read_csv("cifar10_results_2.csv", index_col="epoch"))
plot_angles(pd.read_csv("cifar10_results_3.csv", index_col="epoch"))
plot_angles(pd.read_csv("cifar10_results_4.csv", index_col="epoch"))
plot_angles(pd.read_csv("cifar10_results_5.csv", index_col="epoch"))
```
