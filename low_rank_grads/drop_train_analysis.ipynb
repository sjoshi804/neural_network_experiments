```python
import pandas as pd
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
```

```python
runs = sorted(glob("drop_train_results/modelrun*.csv"))
run_names = [x.split("/")[-1].split(".")[0] for x in runs]
run_names
```

```python
def plot_run(name: str):
    data = pd.read_csv(f"drop_train_results/{name}.csv", index_col="epoch")
    print(name, data["val_categorical_accuracy"].max(), np.argmax(data["val_categorical_accuracy"]))
    plt.plot(data["loss"], label=name)
```

```python
plt.figure(figsize=(16, 9))
for run in run_names:
    plot_run(run)
plt.legend()
plt.show()
```
